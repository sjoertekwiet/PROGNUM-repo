Question 2.15:

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
x=set(night1)
print(f"night1: {x}")
y=set(night2)
print(f"night2: {y}")
z=x.intersection(y)
print(f"Intersection: {z}")
if 5 in z:
    print(f"5 is an element of the intersection list")
else:
    print(f"5 is not an element of the intersection list")
if 5 in x:
    print("5 is in night1")
else:
    print("5 is not in night1")
```

```python
from math import *
a=float(input("a="))
b=float(input("b="))
c=float(input("c="))
D=b**2-4*a*c
x1=((-b+(b**2-4*a*c)**0.5)/2*a)
x2=((-b-(b**2-4*a*c)**0.5)/2*a)
if D>0:
    print(f"The roots are x1={x1} and x2={x2}")
elif D==0:
    print(f"The root is x={x1}")
else:
    print("There are no roots")
```
