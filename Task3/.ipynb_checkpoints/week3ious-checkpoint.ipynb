```python
intensity=[2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0]
total=0
N=0
#total=0 and N=0 are here to tell the loop below to start the sum at 0
for I in intensity:
    if I>0:
        total += I  #This adds up every value in the list
        N += 1      #This just adds 1 for every new value in the list so it counts the number of terms
average=total/N
print(f"Average I: {average}")
x=0 #again starting the sum in the loop at 0
for I in intensity:   #This loop calculates the sum part in the standard deviation
    if I>0:
        x += (I-average)**2 #calculates x_i - average for every i and sums it
sd=(x/N)**0.5 #formula for standard deviation
print(f"      Standard deviation: {sd}")
import numpy as np 
array=np.array(intensity) #creates a numpy array
parray=array[array>0]     #creates a numpy array with only positive values
std=np.std(parray)        #calculates standard deviation using numpy
print(f"Numpy standard deviation: {std}")
```

### Question 3.17

```python
def stdev(y):
    N=len(y)
    mean=np.sum(y)/len(y)
    x=np.sum((y-mean)**2)
    std=np.sqrt(x/N)
    return std
inputarray=np.array([1,1,22,2,2,33,44,55,66,77,88,86,6,4,4])
myresult=stdev(inputarray)
npresult=np.std(inputarray)
print(f"My result:{myresult}, Numpy result:{npresult}")
def oneliner(y):
    return np.sqrt(np.sum((y-np.mean(y))**2)/len(y))
o_l_result=oneliner(inputarray)
print(f"Oneliner result:{o_l_result}")
```
